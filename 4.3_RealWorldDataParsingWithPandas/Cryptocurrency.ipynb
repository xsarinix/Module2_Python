```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "bitcoin_cash_price.csv"
dash_csv = "dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
bitcoin_dash_merge=pd.merge(bitcoin_df, dash_df, on="Date")
bitcoin_dash_merge.head()
```

```python
# Rename columns so that they are differentiated
rename_bitcoin_dash=bitcoin_dash_merge.rename(columns={"Open_x":"Bitcoin Open", 
                                                      "High_x": "Bitcoin High",
                                                       "Low_x": "Bitcoin Low",
                                                       "Close_x": "Bitcoin Close",
                                                       "Volume_x": "Bitcoin Volume",
                                                       "Market Cap_x": "Bitcoin Market Cap",
                                                       "Open_y": "Dash Open",
                                                       "High_y": "Dash High",                                                      
                                                       "Low_y": "Dash Low",
                                                       "Close_y": "Dash Close",
                                                       "Volume_y": "Dash Volume",
                                                       "Market Cap_y": "Dash Market Cap"})
rename_bitcoin_dash.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
bitcoin_dash_merge_suffixes=pd.merge(bitcoin_df,dash_df,
                                     on="Date",suffixes=[" Bitcoin", " Dash"])
bitcoin_dash_merge_suffixes.head()
```

```python
bitcoin_dash_merge_suffixes.describe()
```

```python
# Collecting best open for Bitcoin and Dash
best_open_bitcoin=bitcoin_dash_merge_suffixes["Open Bitcoin"].max()
best_open_dash=bitcoin_dash_merge_suffixes["Open Dash"].max()
print(f"Bitcoin Best Open: {best_open_bitcoin}")
print(f"Das Best Open: {best_open_dash}")

# Collecting best close for Bitcoin and Dash


# Collecting the total volume for Bitcoin and Dash
```
