```python
import pandas as pd
```

```python
# The path to our CSV file
kickstarter_path="KickstarterData.csv"

# Read our Kickstarter data into pandas
kickstarter_data=pd.read_csv(kickstarter_path)
kickstarter_data.head()
```

```python
# Get a list of all of our columns for easy reference
kickstarter_data.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
kickstarter_clean1=kickstarter_data[["name","goal","pledged", "state", "country", "staff_pick", "backers_count", "spotlight"]]
kickstarter_clean1.head(100)
```

```python
# Remove projects that made no money at all
kickstarter_clean1["pledged"]=pd.to_numeric(kickstarter_clean1["pledged"])
kickstarter_clean2=kickstarter_clean1.loc[kickstarter_clean1["pledged"] > 0,:]
kickstarter_clean2.head()
```

```python
# Collect only those projects that were hosted in the US
kickstarter_clean_over0_US_only=kickstarter_clean2.loc[kickstarter_clean2["country"] == "US",:]
kickstarter_clean_over0_US_only.head()
```

```python
# Create a new column that finds the average amount pledged to a project
avg_pledge=kickstarter_clean_over0_US_only["pledged"]/kickstarter_clean_over0_US_only["backers_count"]
kickstarter_over0_US_avgpledge=kickstarter_clean_over0_US_only
kickstarter_over0_US_avgpledge["Avg Pledge"]=avg_pledge
kickstarter_over0_US_avgpledge.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float
kickstarter_over0_US_avgpledge[["goal","pledged","Avg Pledge"]]=pd.to_numeric(kickstarter_over0_US_avgpledge[["goal","pledged","Avg Pledge"]])
# Then Format to go to two decimal places, include a dollar sign, and use comma notation
```
