```python
import pandas as pd
```

```python
# The path to our CSV file
kickstarter_path="KickstarterData.csv"

# Read our Kickstarter data into pandas
kickstarter_data=pd.read_csv(kickstarter_path)
kickstarter_data.head()
```

```python
# Get a list of all of our columns for easy reference
kickstarter_data.columns
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
kickstarter_trim=kickstarter_data[["name","goal","pledged", "state", "country", "staff_pick", "backers_count", "spotlight"]].sort_values("pledged")
kickstarter_trim.head()
```

```python
# Remove projects that made no money at all

# solution using a filter to hide $0 pledged projects 
#kickstarter_clean2=kickstarter_clean1.loc[kickstarter_clean1["pledged"] > 0,:]
#kickstarter_clean2.head()

# solution using drop to actually DELETE rows less than 0
kickstarter_over0=kickstarter_trim
kickstarter_over0=kickstarter_over0.drop(kickstarter_over0[kickstarter_over0.pledged < 1].index)
kickstarter_over0.sort_values("country").head()
```

```python
# Collect only those projects that were hosted in the US
# solution using a filter to hide non-US projects
#kickstarter_over0_US=kickstarter_over0.loc[kickstarter_over0["country"] == "US",:]
#kickstarter_over0_US.head()

#solution using drop to delete non-US rows
kickstarter_over0_us=kickstarter_over0
kickstarter_over0_us=kickstarter_over0_us.drop(kickstarter_over0_us[kickstarter_over0_us.country != "US"].index)
kickstarter_over0_us.head()
```

```python
# Create a new column that finds the average amount pledged to a project
avg_pledge=kickstarter_over0_us["pledged"]/kickstarter_over0_us["backers_count"]
kickstarter_over0_us_avgpledge=kickstarter_over0_us.sort_index().reset_index(drop=True)
kickstarter_over0_us_avgpledge["Avg Pledge"]=avg_pledge
kickstarter_over0_us_avgpledge.head()
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float

#data was already in the proper forms, so convert to objects then convert back
kickstarter_over0_us_avgpledge.to_object(kickstarter_over0_us_avgpledge["Avg Pledge"])
# kickstarter_over0_us_avgpledge.dtypes
#kickstarter_over0_US_avgpledge["goal"]=pd.to_numeric(kickstarter_over0_US_avgpledge["goal"])
# Then Format to go to two decimal places, include a dollar sign, and use comma notation
```
