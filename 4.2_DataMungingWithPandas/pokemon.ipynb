```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# File to load
file="pokemon.csv"
```

```python
# Read with Pandas
pokedex_full=pd.read_csv(file)
pokedex_full.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
pokedex2=pokedex_full[
    ['Type 1', 'HP', 'Attack', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokedex=pokedex2.rename(columns={"Type 1":"Type"})
pokedex.head()
```

```python
 # Create a dataframe of the average stats for each type of pokemon.
grouped_pokedex=pokedex.groupby(["Type"])
grouped_pokedex.mean().head()
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.
pokedex["Total"]=pokedex["HP"]+pokedex["Attack"]+pokedex["Sp. Atk"]+pokedex["Sp. Def"]+pokedex["Speed"]
grouped_pokedex.mean().head()
```
