```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Create reference to CSV file
csv_path = "Soccer2018Data.csv"

# Import the CSV into a pandas DataFrame
soccer_2018_df = pd.read_csv(csv_path, low_memory=False)
soccer_2018_df.head()
```

```python
soccer_2018_df.count()
```

```python
soccer_drop=soccer_2018_df.dropna(how='any')
soccer_drop.count()
```

```python
# Collect a list of all the unique values in "Preferred Position"
soccer_drop["Preferred Position"].value_counts()
```

```python
# Looking only at strikers (ST) to start
strikers=soccer_drop.loc[soccer_drop["Preferred Position"]=="ST",:]
strikers.head()
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
strikers_st=strikers.sort_values("ST", ascending=True).reset_index(drop=True)
# Reset the index so that the index is now based on the sorting locations
strikers_st.head()
```

```python
# Save all of the information collected on the worst striker
worst_striker=strikers_st.iloc[0,:]
print(worst_striker)
```
