```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
file="crime_incident_data2017.csv"

# Import the data into a Pandas DataFrame
crime=pd.read_csv(file)
crime.head()
```

```python
# look for missing values
crime.count()
```

```python
# drop null rows
crime_drop=crime.dropna(how='any')
```

```python
# verify counts
crime.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
crime_drop['Offense Type'].value_counts()
```

```python
# Combine similar offenses
crime_drop["Offense Type"]=crime_drop["Offense Type"].replace(
    {"Assisting or Promoting Prostitution":"Prostitution",
     "Commercial Sex Acts":"Prostitution", 
     "Drug Equipment Charges":"Drug/Narcotic Violations"}
)
crime_drop["Offense Type"].value_counts()
```

```python
# Create a new DataFrame that looks into a specific neighborhood
centennial=crime_drop.loc[crime_drop["Neighborhood"] == "Centennial",:]
centennial.head()
```

```python
centennial.to_excel("centennial.xlsx", index=False, header=True)
```
