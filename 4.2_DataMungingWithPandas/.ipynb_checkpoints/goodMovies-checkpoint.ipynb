```python
# Dependencies
import pandas as pd
```

```python
# Load in File from resources
file = "movie_scores.csv"
```

```python
# Read and display with pandas
original=pd.read_csv(file)
original.head()
```

```python
# List all the columns the table provides
original.columns
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB
imdb=original[["FILM","IMDB","IMDB_norm","IMDB_norm_round","IMDB_user_vote_count"]]
imdb.head()
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
over7=imdb.loc[imdb["IMDB"] > 7,:]
over7
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
under20=imdb.loc[imdb["IMDB_user_vote_count"]<20000]
under20
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
imdb.to_csv("imdb.csv", index=False, header=True)
```

```python
#Export to excel file
imdb.to_excel("imdb.xlsx",index=False, header=True)
```
