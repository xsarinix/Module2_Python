```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
# bar graph
gym_x=np.arange(0, len(gyms))
plt.bar(gym_x, members, facecolor="red", alpha=0.5, align="center")
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
# line graph
plt.plot(x_axis, sin)
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
# pie chart
plt.pie(members, explode=explode, labels=gyms, colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
```
