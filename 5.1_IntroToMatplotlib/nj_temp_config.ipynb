```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define x-axis and fahrenheit list
fahrenheit=[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
x_axis=np.arange(1,13,1)
x_axis
```

```python
# plot fahrenheit
plt.plot(x_axis, fahrenheit)
plt.xlabel("Months")
plt.ylabel("Degrees")
```

```python
# convert fahrenheit to celsius
celsius=[(f-32)*0.56 for f in fahrenheit]
```

```python
plt.plot(x_axis, celsius)
plt.xlabel("Months")
plt.ylabel("Degrees")
```

```python
fahrenheit_handle, = plt.plot(x_axis, fahrenheit, marker="+", color="blue", label="Fahrenheit")
celsius_handle, = plt.plot(x_axis, celsius, marker="*", color="red", label="Celsius")
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.legend(loc="lower right")
plt.savefig("nj_temp_config.png")
plt.show()
```
